#Silver Layer walkthrough

#Schema overwriting from bronze to silver

```python
df_silver.write.mode("overwrite").option("overwriteSchema", "true").format("delta").saveAsTable("workspace.default.road_accidents_dataset")
display(df_silver)
df_silver.printSchema()
df_silver.show(5)
```
